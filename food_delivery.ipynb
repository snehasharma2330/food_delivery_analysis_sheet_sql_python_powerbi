```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("food_delivery_cleaned - Sheet1.csv") 
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Rating'].value_counts()
```

```python
df.groupby('Food_Category')['Rating'].mean().sort_values(ascending=False)
```

```python
df[['Delivery_Time', 'Rating']].corr()
```

```python
df.nlargest(5, 'Order_Amount')
```

```python
df.groupby('Restaurant')['Order_Amount'].sum().sort_values(ascending=False).plot(kind= 'bar')
plt.title('Revenue by Restaurant')
plt.xlabel('Restaurant')
plt.ylabel('Revenue')

plt.show()
```

```python
df['Payment_Method'].value_counts().plot(kind ='pie', autopct = '%1.1f%%')
plt.title('Payment Method Distribution')
plt.ylabel("")
plt.show()
```

```python
df['Order_Amount'].plot(kind ='hist', bins =5)
plt.title('Revenue Distributed')
plt.xlabel('Revenue')

plt.show()
```

```python
plt.scatter(df['Delivery_Time'], df['Order_Amount'])
plt.title('Delivery time vs Revenue')
plt.xlabel('Delivery time')
plt.ylabel('Revenue')

plt.show()


```
